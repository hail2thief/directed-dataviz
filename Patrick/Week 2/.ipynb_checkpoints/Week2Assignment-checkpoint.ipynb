Set-up.

```r
# Install necessary packages
install.packages(c('ggplot2', 'readxl', 'tidyverse'))
library('ggplot2')
library('readxl')
library('tidyverse')
```

```r
# Read in data
df <- read_excel('Department_Ideology_Scores.xlsx')

# Remove rows containing NA values
df <- df[complete.cases(df),]

# Creating a factor variable for agency
df$agcy_name <- as.factor(df$agcy_name)

# Grouping
df1 <- df %>%
  group_by(agcy_name)

head(df)
```

Creating base plot.

```r
z <- ggplot(data = df1, aes(x = agcy_name, y = ideology))
z
```
