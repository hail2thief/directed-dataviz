Data Visualization

- make a barplot, stacked barplot, heat map, nested area plot

```r
# Load in data and install necessary packages:
install.packages("reshape2")
library("reshape2")
library("ggplot2")

# The data I am using for this assignment is the National Material Capabilites dataset hosted by the Correlates of War (cow)
original.df <- read.csv("NMC-60-wsupplementary.csv")

# Filtering out countries to create dataset of Great Power observations only
data <- original.df[original.df$statenme %in% c("United States of America", "Russia", "China"), ]
```

First, lets create a barplot of the military expenditures for each country in each year represented in the dataset.

```r
# Create barplot for China
china_data <- data[data$statenme == "China", ]

p_china <- ggplot(china_data, aes(x = year, y = milex)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  labs(title = "Military Expenditure for China",
       x = "Year",
       y = "Military Expenditure") +
  theme_minimal()

print(p_china)

# Create barplot for the US
us_data <- data[data$statenme == "United States of America", ]

p_us <- ggplot(us_data, aes(x = year, y = milex)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  labs(title = "Military Expenditure for US",
       x = "Year",
       y = "Military Expenditure") +
  theme_minimal()

print(p_us)

# Create barplot for Russia
russia_data <- data[data$statenme == "Russia", ]

p_russia <- ggplot(russia_data, aes(x = year, y = milex)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  labs(title = "Military Expenditure for Russia",
       x = "Year",
       y = "Military Expenditure") +
  theme_minimal()

print(p_russia)
```

Next, let's stack these plots to create a stacked barplot that demonstrates all the information in a single plot.

```r
p_stacked <- ggplot(data, aes(x = year, y = milex, fill = statenme)) +
  geom_bar(stat = "identity") +
  labs(title = "Military Expenditure Over Time",
       x = "Year",
       y = "Military Expenditure") +
  theme_minimal()

print(p_stacked)
```

Next, it's time to create a heatmap that illustrates a correlation matrix of the following variables found within the National Material Capabilities dataset.
- (irst) Iron and Steel Production
- (milex) Military Expenditures
- (milper) Military Personnel
- (tpop) Total population size
- (upop) Urban population size (population living in cities with population greater than 100,000; in thousands)

```r
install.packages("reshape2")
library(reshape2)
```

```r
# Select relevant metrics and compute correlation matrix
metrics <- data[, c("irst", "milex", "milper", "tpop", "upop")]
cor_matrix <- cor(metrics, use = "complete.obs")  # use complete.obs ignores missing values

# Melt the correlation matrix for ggplot2
melted_cor_matrix <- melt(cor_matrix)

# Create the heatmap
p_heatmap <- ggplot(melted_cor_matrix, aes(x = Var1, y = Var2, fill = value)) +
  geom_tile() + 
  scale_fill_gradient2(low = "blue", high = "red", mid = "white", 
                       midpoint = 0, limit = c(-1,1), space = "Lab",
                       name="Correlation") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  labs(title = "Correlation Heatmap of Military Metrics")

print(p_heatmap)
```

Finally, let's create a nested or stacked area plot to illustrate trends in military expenditure in Russia, China, and the United States.

```r
# Using ggplot2 for a stacked area plot
ggplot(data, aes(x = year, y = irst, fill = stateabb, group = stateabb)) +
  geom_area(alpha = 0.6, position = 'stack') +
  labs(title = "Iron and Steel Production Over Time",
       x = "Year",
       y = "Production") +
  theme_minimal()
```
